```python
def fillpanel(wire,panel=[],x=0,y=0):
    if len(wire)==0:
        return panel
    segment=wire[0]
    direction=segment[0]
    length=int(segment[1:])
    if direction=='R':
        for n in range(length):
            panel.append((x,y+1+n))
        return fillpanel(wire[1:],panel,x,y+length)
    elif direction=='L':
        for n in range(length):
            panel.append((x,y-n))
        return fillpanel(wire[1:],panel,x,y-length)    
    elif direction=='U':
        for n in range(length):
            panel.append((x+1+n,y))
        return fillpanel(wire[1:],panel,x+length,y)
    elif direction=='D':
        for n in range(length):
            panel.append((x-n-1,y))
        return fillpanel(wire[1:],panel,x-length,y)
    else:
        return (panel,wire,x,y)
```
