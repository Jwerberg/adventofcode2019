```python
def parameters(n):
    return [int(n/10000),int((n-int(n/10000)*10000)/1000),int((n-int(n/1000)*1000)/100)][::-1]
```

```python
def opcode(l,offset=0,input_value=0):
    code=l[0+offset]
    basecode=int(str(code)[-2:])
    p=parameters(code)
    if basecode in (1,2,3,4,5,6,7,8):
        if p[0]==1:
            p0=1+offset
        else:
            p0=l[1+offset]
    if basecode in (1,2,5,6,7,8):
        if p[1]==1:
            p1=2+offset
        else:
            p1=l[2+offset]
        if p[2]==1:
            p2=3+offset
        else:
            p2=l[3+offset]
    if basecode==99:
        print('HALT')
        return l
    elif basecode==1:
        l[p2]=l[p0]+l[p1]
        return opcode(l,offset+4) 
    elif basecode==2:
        l[p2]=l[p0]*l[p1]
        return opcode(l,offset+4)
    elif basecode==3:
        l[p0]=input_value
        return opcode(l,offset+2)
    elif basecode==4:
        output=l[p0]
        print('OUTPUT',output)
        return opcode(l,offset+2)   
    elif basecode==5:
        if l[p0]!=0:
            return opcode(l,l[p1])
        else:
            return opcode(l,offset+3)
    elif basecode==6:
        if l[p0]==0:
            return opcode(l,l[p1])
        else:
            return opcode(l,offset+3)
    elif basecode==7:
        if l[p0]<l[p1]:
            l[p2]=1
        else:
            l[p2]=0
        return opcode(l,offset+4)
    elif basecode==8:
        if l[p0]==l[p1]:
            l[p2]=1
        else:
            l[p2]=0
        return opcode(l,offset+4)    
    else:
        print('SOMETHING UNEXPECTED',offset,code,l[offset],l[offset+1])
        return(l)
```
